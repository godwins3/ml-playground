```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

pd.set_option('display.max.columns', None)
print('setup complete')
```

```python
# get the html page
page = requests.get("https://www.must.ac.ke/kuccps-2020-admission-list/", verify = False)
soup = BeautifulSoup(page.text, 'lxml')
soup
```

```python
# get the table

table = soup.find('table')
print(table)
headers = []
```

```python
for rows in table.find_all('tr')[:1]:
    data = rows.find_all('td')
    titles = [td.text.strip() for td in data]
    headers.append(titles)
df =pd.DataFrame(columns = headers)
df
```

```python
# getting the table content

for row in table.find_all('tr')[1:]:
    row_data = row.find_all('td')
    content = [td.text for td in row_data]
    length = len(df)
    df.loc[length]=content

df
```
